```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('titanic-passengers.csv', sep=';', encoding = "ISO-8859-1")
df.head()
```

```python
#faire la correlation entre sexe et age
correlation = df1['Sex'].str.get_dummies(sep=' ').corrwith(df1['Age'])
plt.ylabel('Correlation en Age et Sex')
plt.plot(correlation)
plt.show()
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df1)
```
